```python
# load library  env name applied_ai
import os
import pandas as pd
import numpy as np
import string
from nltk.corpus import stopwords
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# featur e
from sklearn.feature_extraction.text import CountVectorizer
```

```python
review_dataset= pd.read_csv(r'D:\new_ai\NLP_basic_to_advanced\IMDB Dataset.csv')[:10000]
print(f"{'---' * 15} Dataset Head {'---' * 15}")
print(review_dataset.head())

print(f"{'---'*15} Dataset Size {'---'*15}")
print(f"Dataset Size {review_dataset.shape}")
#print(review_dataset.info())

print(review_dataset['sentiment'].value_counts())
print("---" * 50)
print(review_dataset.isnull().sum())
```

```python
print(review_dataset.duplicated().sum())
review_dataset.drop_duplicates(inplace=True)
print(f"Dataset Size {review_dataset.shape}")
print(review_dataset.info())
print(review_dataset['sentiment'].value_counts())
print(review_dataset.isnull().sum())
```

```python
# HTML Tag remove
import re

def remove_html(input_text):
   clean_text = re.sub(re.compile('<.*?>'),'', input_text)
   return clean_text
```

```python
# remove punctuation 
punc = string.punctuation
print(punc) 


def remove_punctuation(input_text):
   table= str.maketrans('','',punc)
   removed= input_text.translate(table)
   return removed
```

```python
   #stop word removal 
sw_list = stopwords.words('english')

def remove_stopwords(input_text):
      words= input_text.split()
      filtered_words=[]
      # filtered_words = [word for word in words if word.lower() not in sw_list]
      for word in words:
            if word.lower() not in sw_list:
               filtered_words.append(word)
      return " ".join(filtered_words)
         
         
```

```python
# apply on entire ds 

review_dataset['review'] = review_dataset['review'].apply(remove_html)
review_dataset['review']= review_dataset['review'].str.lower()
review_dataset['review']=review_dataset['review'].apply(remove_punctuation)
print(review_dataset.head())
```

```python
X= review_dataset['review']

y=review_dataset['sentiment']

print(f"(X shape is {X.shape}, y shape is {y.shape})")
```

```python
le = LabelEncoder()
y= le.fit_transform(y)
print(y[1:50])
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y, test_size=0.3, random_state=42)

print(f"Train Shape is {X_train.shape[0]}\n")
print(f"Test Shape is {X_test.shape[0]}\n")
print(f" Train Label shape {y_train.shape[0]}\n")
```

```python
# featur extraction 
feat = CountVectorizer()  # BoW 
X_train_feat= feat.fit_transform(X_train)
print(feat.get_feature_names_out()[1:50])

X_test_feat= feat.transform(X_test)
print(feat.get_feature_names_out()[1:50])
```

```python
X_train_dense = X_train_feat.toarray()
X_test_dense= X_test_feat.toarray()
```

```python
# classifiers 
from sklearn.naive_bayes import GaussianNB
clf_gnb= GaussianNB()
clf_gnb.fit(X_train_dense,y_train)

y_pred = clf_gnb.predict(X_test_dense)
#print(y_pred)
print(f"Accuracy is {np.mean(y_test==y_pred)}")
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
# MN-GB
mngb= MultinomialNB()
mngb.fit(X_train_dense,y_train)

y_pred_mngb = mngb.predict(X_test_dense)
#print(y_pred)
print(f"Accuracy using multinomia is {np.mean(y_test==y_pred_mngb)}")

#LR

lr= LogisticRegression()
lr.fit(X_train_dense,y_train)

y_pred_lr = lr.predict(X_test_dense)
#print(y_pred)
print(f"Accuracy using Logistic Regression is {np.mean(y_test==y_pred_lr)}")


#SVM

sv= LinearSVC()
sv.fit(X_train_dense,y_train)

y_pred_sv = sv.predict(X_test_dense)
#print(y_pred)
print(f"Accuracy using SVM is {np.mean(y_test==y_pred_sv)}")
```

```python
review_test= ["Movie was ok but no so much."]
review_feat = feat.transform(review_test)
prediction = lr.predict(review_feat)
print(le.inverse_transform(prediction))
```
