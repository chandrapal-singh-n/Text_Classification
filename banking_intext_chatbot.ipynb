```python
import os 
import numpy as np
import pandas as pd


import string
from nltk.corpus import stopwords
```

```python
train_dataset_path = r"D:\new_ai\banking77\train.csv"
test_dataset_path= r"D:\new_ai\banking77\test.csv"
vali_dataset_path = r"D:\new_ai\banking77\dev.csv"
```

```python
train_data= pd.read_csv(train_dataset_path)
test_data= pd.read_csv(test_dataset_path)
val_data= pd.read_csv(vali_dataset_path)
print(f" {'_'*30}  Training Head  {'_'*30}")
print(train_data.head())
print(f" {'_'*30}  Testing Head  {'_'*30}")
print(test_data.head())
print(f" {'_'*30}  Validation Head  {'_'*30}")
print(val_data.head())
```

```python
print(train_data.describe())
print(train_data.isnull().sum())
print(train_data['label'].value_counts())
```

```python

# HTML Tag remove
import re

def remove_html(input_text):
   clean_text = re.sub(re.compile('<.*?>'),'', input_text)
   return clean_text
# remove punctuation 
punc = string.punctuation
print(punc) 


def remove_punctuation(input_text):
   table= str.maketrans('','',punc)
   removed= input_text.translate(table)
   return removed


   #stop word removal 
sw_list = stopwords.words('english')

def remove_stopwords(input_text):
      words= input_text.split()
      filtered_words=[]
      # filtered_words = [word for word in words if word.lower() not in sw_list]
      for word in words:
            if word.lower() not in sw_list:
               filtered_words.append(word)
      return " ".join(filtered_words)
         
         
```

```python
train_data['text'] = train_data['text'].apply(remove_html)
train_data['text']= train_data['text'].str.lower()
train_data['text']=train_data['text'].apply(remove_punctuation)

train_data['text'] = train_data['text'].apply(remove_stopwords)
print(train_data.head())


test_data['text'] = test_data['text'].apply(remove_html)
test_data['text']= test_data['text'].str.lower()
test_data['text']=test_data['text'].apply(remove_punctuation)
test_data['text'] = test_data['text'].apply(remove_stopwords)
print(train_data.head())


val_data['text'] = val_data['text'].apply(remove_html)
val_data['text']= val_data['text'].str.lower()
val_data['text']=val_data['text'].apply(remove_punctuation)
val_data['text'] = val_data['text'].apply(remove_stopwords)
print(train_data.head())




```

```python
X_train= train_data['text']
y_train=train_data['label']


X_test= test_data['text']
y_test=test_data['label']



X_val= val_data['text']
y_val=val_data['label']
print(f"(X shape is {X_train.shape}, y shape is {y_train.shape})")
print(f"(X shape is {X_test.shape}, y shape is {y_test.shape})")
print(f"(X shape is {X_val.shape}, y shape is {y_val.shape})")
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
y_train= le.fit_transform(y_train)
y_test= le.transform(y_test)
y_val= le.transform(y_val)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# featur extraction 
#feat = TfidfVectorizer()  # BoW 

feat = TfidfVectorizer(
    ngram_range=(1,2),
    min_df=2,
    max_df=0.95
)
X_train_feat= feat.fit_transform(X_train)
print(feat.get_feature_names_out()[1:50])

X_test_feat= feat.transform(X_test)
print(feat.get_feature_names_out()[1:50])

X_val_feat= feat.transform(X_val)
print(feat.get_feature_names_out()[1:50])
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
# MN-GB
mngb= MultinomialNB()
mngb.fit(X_train_feat,y_train)

y_pred_mngb = mngb.predict(X_test_feat)
#print(y_pred)
print(f"Accuracy using multinomia is {np.mean(y_test==y_pred_mngb)}")

#LR

#lr= LogisticRegression()
lr = LogisticRegression(
    C=5,
    max_iter=2000,
    n_jobs=-1
)
lr.fit(X_train_feat,y_train)

y_pred_lr = lr.predict(X_test_feat)
#print(y_pred)
print(f"Accuracy using Logistic Regression is {np.mean(y_test==y_pred_lr)}")


#SVM

# sv= LinearSVC()
# sv.fit(X_train_dense,y_train)

# y_pred_sv = sv.predict(X_test_dense)
# #print(y_pred)
# print(f"Accuracy using SVM is {np.mean(y_test==y_pred_sv)}")
```

```python
from sklearn.metrics import classification_report

print(classification_report(
    y_test,
    y_pred_lr
))
```
